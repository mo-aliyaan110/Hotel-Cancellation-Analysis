# Data Analysis and Visualizations

```python
cancelled_perc = df['is_canceled'].value_counts(normalize=True)
print(cancelled_perc)

plt.figure(figsize=(5,4))
plt.title('Count of cancellation status')
plt.bar(['Not Cancelled', 'Cancelled'], df['is_canceled'].value_counts(), edgecolor='k', width=0.7)
plt.show()
```

```python
df.head(2)
```

```python
plt.figure(figsize=(8,4))
ax1=sns.countplot(x='hotel', hue='is_canceled', data=df, palette='Greens')
# legend_labels = ax1.get_legend_handles_labels()
# ax1.legend(bbox_to_anchor(1,1))
plt.title('Reservation status in different hotels', size=20)
plt.xlabel('hotel')
plt.ylabel('number of reservations')
plt.legend(fontsize=15)
plt.show()
```

```python
# Now let's see the cancellation percentage of both hotels types
```

```python
df.head(2)
```

```python
resort_hotel = df[df['hotel']=='Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
city_hotel = df[df['hotel']=='City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)
```

```python
df.head(3)
```

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
# Now its time to plot the results.
```

```python
plt.figure(figsize=(20,8))
plt.title('Average Daily Rate in City and Resort Hotel', size='20')
plt.plot(resort_hotel.index, resort_hotel['adr'], label='Resort Hotel')
plt.plot(city_hotel.index, city_hotel['adr'], label='City Hotel')
plt.legend(fontsize=20)
plt.show()
```

```python
# Now we need to check on what month, there were more reservations and which month had the more cancellations.
```

```python
df.head(3)
```

```python
df['month'] = df['reservation_status_date'].dt.month

plt.figure(figsize=(16, 8))
plt.title('Reservation Status Per Month.', size=20)
sns.countplot(x='month', hue='is_canceled', data=df,  palette='bright' )
plt.xlabel('months')
plt.ylabel('number of reservations')
plt.legend(['not_cancelled', 'cancelled'],fontsize=10)
plt.show()
```

```python
# Now we shall see that if our cancellation rate have any effect on the adr, we shall use now bar plot as we have only one class in the hue
```

```python
plt.figure(figsize=(15,8))

plt.title("Average daily rate of each month", size=20)

sns.barplot('month', 'adr', data=df[df['is_canceled']==1].groupby('month')[['adr']].sum().reset_index())
plt.xlabel('Months')
plt.ylabel('Average Daily Rate')
plt.show()
```

```python
# Now we shall see, which country has higher cancellation rates. 
```

```python
cancelled = df[df['is_canceled']==1]

top_10_countries = cancelled['country'].value_counts()[:10]
plt.figure(figsize=(10,8))

plt.title('Cancellation rates in Top 10 Countries', size=20)
plt.pie(top_10_countries, autopct='%.2f', labels= top_10_countries.index)
plt.show()
```
